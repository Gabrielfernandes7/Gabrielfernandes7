```python
# dataset salary
import pandas as pd

# Carregar o dataset
data_salary = pd.read_csv('/content/sample_data/Salary_dataset.csv')

data_salary.head(15)
```

```python
# Informações descritivas do dataset
data_salary.describe()
```

```python
# Colunas dos dados
print('Colunas')
data_salary.columns
```

```python
# dropna(axis=0) remove linhas do dataset salary_dataset.csv que possuem valores ausentes
filtered_data_salary = data_salary.dropna(axis=0)
```

```python
# Selecione uma meta de previsão (prediction target)
y = filtered_data_salary.Salary

# Selecionar a feature
salary_feature = ['YearsExperience']
x = data_salary[salary_feature]

# describe() nos atributos
x.describe()
```

```python
x.head(10)
```

# Construindo e utilizando um model

### Os passos para construir utilizar o modelo são:

* Passo 1: Definir modelo árvore de decisão (DecisionTreeRegressor), alguns outros tipos de modelo

* Passo 2: Treinamento (**Fit**) serve para captura padrões dos dados fornecidos

* Passo 3: Predição ou previsão, como se diz na expressão inglesa "just what it sounds like"

* Passo 4: Avaliar (Evaluate) a precisão dos modelos

```python
from sklearn.tree import DecisionTreeRegressor

# Árvore de decisão regressor
salary_model = DecisionTreeRegressor(random_state=1)

salary_model.fit(x, y)
```

```python
print('Fazendo as previsões de 5 pessoas de seus salários de acordo com os anos de esperiência \n')

print(x.head())

print('\n')

predict_salary = salary_model.predict(x)

print(f'Previsão do salário:\n {predict_salary}', end=' ')
```

# Model validation

### Validação do modelo:
Nessa lição você irá mensurar a qualidade da previsão do modelo. Mensurar a qualidade do modelo é a chave para iterativamente melhorar seus modelos.

Existem muitas métricas que servem para avaliar nossos modelos , mas nós iremos começar com uma chama da MAE (Mean Absolute Error)

```python
from sklearn.metrics import mean_absolute_error

print('A Média do erro absolute é:\n ')
mean_absolute_error(y, predict_salary)
```

# O prolema das pountuações "in-sample"

O problema se encontra em utilizarmos uma única amostra para definir a precisão de nosso modelo

```python
from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)

salary_model = DecisionTreeRegressor()
salary_model.fit(train_x, train_y)

val_predict = salary_model.predict(val_x)

print('A média do erro absoluto é:\n ')
print(mean_absolute_error(val_y, val_predict))
```
